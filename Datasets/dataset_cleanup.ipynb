```python
import pandas as pd
import numpy as np
from ucimlrepo import fetch_ucirepo
import sklearn.datasets as datasets
```

```python
#for datasets taken from UCI machine learning datasets
def to_pandas_first_files(dataset_):
  X = dataset_.data.features
  y = dataset_.data.targets

  data_f = pd.DataFrame(X)
  data_f["targets"] = y
  return data_f


#for datasests taken from Sklearn.datasets
def to_pandas_second_files(dataset_):
  X = dataset_.data
  y = dataset_.target
  data_f = pd.DataFrame(X)
  data_f["targets"] = y
  return data_f
```

# ISOLATE DATASET

```python
isolet = fetch_ucirepo(id=54)
isolet_df = to_pandas_first_files(isolet)
isolet_df.shape
```

# MUSK_DATASET_V2

```python
musk_version_2 = fetch_ucirepo(id=75)
musk_version_2_df = to_pandas_first_files(musk_version_2)
musk_version_2_df.shape
```
